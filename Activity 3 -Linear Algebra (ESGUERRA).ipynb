### _TASK 1_ 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

$$NV = a*vect_1 + b*vect_2 $$$$v_1 = \begin{bmatrix} 2\\8 \\\end{bmatrix} , v_2 = \begin{bmatrix} -7\\4 \\\end{bmatrix} $$

```python
vect1 = np.array([2,8])
vect2 = np.array([-7,4])
a = 1
b = 2 
NV = a*vect1 + b*vect2
print(NV)
```

```python
origin = np.zeros(NV.shape)

plt.figure(figsize=(5,5))
plt.quiver(*origin, *NV, color=['r'], scale=1, units='xy')
plt.quiver(*origin, *vect1, color=['g'], scale=1, units='xy')
plt.quiver(*origin, *vect2, color=['b'], scale=1, units='xy')

plt.grid()
plt.xlim(-20,20)
plt.ylim(-20,20)
plt.text(vect1[0], vect1[1], 'vect1')
plt.text(vect1[0], vect2[1], 'vect2')
plt.text(NV[0], NV[1], 'Linear combination of a*vect1 and b*vect2')
plt.gca().set_aspect('equal')
plt.show()
```

### _TASK 2_

$$Y= \begin{Bmatrix} v_1 \cdot\begin{bmatrix} 2\\8\\\end{bmatrix}, 
v_2 \cdot \begin{bmatrix} -7\\4 \\\end{bmatrix}\end{Bmatrix} $$$$v_1 = \begin{bmatrix} 2\\8\\\end{bmatrix} , v_2 = \begin{bmatrix} -7\\4\\\end{bmatrix} $$

```python
vectA = np.array([2,8])
vectB = np.array([-7,4])
R = np.arange(-10,10,1)
v1, v2 = np.meshgrid(R,R)
vectR = vectA + vectB
spanRx = v1*vectA[0] + v2*vectB[0]
spanRy = v1*vectA[1] + v2*vectB[1]
plt.scatter(spanRx,spanRy, s=5, alpha=0.75)

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.show()
```

$$Z= \begin{Bmatrix} g_1 \cdot\begin{bmatrix} 4\\-3 \\\end{bmatrix}, 
g_2 \cdot \begin{bmatrix} -2\\1 \\\end{bmatrix}\end{Bmatrix} $$$$g_1 = \begin{bmatrix} 4\\-3\\\end{bmatrix} , g_2 = \begin{bmatrix} -2\\1\\\end{bmatrix} $$

```python
vectA = np.array([4,-3])
vectB = np.array([-2,1])
R = np.arange(-10,10,1)
g1, g2 = np.meshgrid(R,R)
vectR = vectA + vectB
spanRx = g1*vectA[1] + g2*vectB[1]
spanRy = g1*vectA[0] + g2*vectB[0]
plt.scatter(spanRx,spanRy, s=5, alpha=0.75)

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.show()
```
